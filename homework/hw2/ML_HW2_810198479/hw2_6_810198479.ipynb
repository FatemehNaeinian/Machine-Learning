```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def data_generator(r0,r1,n):
    lst_x = []
    lst_y = []
    for i in range(n):
        r = random.uniform(r0,r1);
        theta = random.uniform(0,360);
        x = r * np.cos(theta) + 1.5
        y = r * np.sin(theta)
        lst_x.append(x)
        lst_y.append(y)
    return lst_x , lst_y

def normal_generator(mux,muy,sigma,n):
    lst_x = []
    lst_y = []
    for i in range(n):
        x = np.random.normal(mux, sigma)
        y = np.random.normal(muy, sigma)
        lst_x.append(x)
        lst_y.append(y)
    return lst_x , lst_y
    
```

```python

point4_9_x , point4_9_y = data_generator(4,9,200)
point0_6_x , point0_6_y = data_generator(0,6,200)
point_mean01_x , point_mean01_y = normal_generator(1,0,1,100)
point2_6_x , point2_6_y = data_generator(2,6,200)
```

```python
## alef
plt.scatter(point4_9_x , point4_9_y,  c='b')
plt.scatter(point0_6_x , point0_6_y,  c='r')
```

```python
plt.scatter(point_mean01_x , point_mean01_y,  c='g')
plt.scatter(point2_6_x , point2_6_y,  c='purple')
```
