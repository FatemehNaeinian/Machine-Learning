```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from scipy import stats
import seaborn as sns; sns.set()
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.svm import SVC
from sklearn import preprocessing
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import LinearSVC
```

```python
df = pd.read_csv('q5.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
X = df
y = df['target_class']
del X['target_class']
#X = preprocessing.normalize(X)
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
### RBF
C = [1,100,1000]
accuracy_train = []
accuracy_test = []
for c in C:
    svm_clf = Pipeline([
            ("scaler", StandardScaler()),
            ("svm_clf", SVC(kernel="rbf", C=c))
        ])
    svm_clf.fit(X_train, y_train)
    Ypred_train = svm_clf.predict(X_train)
    Ypred_test = svm_clf.predict(X_test)
    accuracy_train.append(metrics.accuracy_score(y_train, Ypred_train))
    accuracy_test.append(metrics.accuracy_score(y_test, Ypred_test))

accuracy_train
```

```python
### Linear
C = [1,100,1000]
accuracy_train = []
accuracy_test = []
for c in C:
    svm_clf = Pipeline([
            ("scaler", StandardScaler()),
            ("svm_clf", SVC(kernel="linear", C=c))
        ])
    svm_clf.fit(X_train, y_train)
    Ypred_train = svm_clf.predict(X_train)
    Ypred_test = svm_clf.predict(X_test)
    accuracy_train.append(metrics.accuracy_score(y_train, Ypred_train))
    accuracy_test.append(metrics.accuracy_score(y_test, Ypred_test))
    
accuracy_train
```
